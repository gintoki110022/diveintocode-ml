```python
from matplotlib import pyplot as plt
%matplotlib inline

import numpy as np
csv_path = r"C:\Users\Admin\.ipynb_checkpoints\mtfuji_data.csv" # Specify the file name (path)

np.set_printoptions(suppress=True) # Set the prohibition of exponential notation
fuji = np.loadtxt(csv_path, delimiter=",", skiprows=1)
```

### [Problem 1] Visualization of data

```python
plt.title("Mt. Fuji")
plt.xlabel("position")
plt.ylabel("elevation[m]")
plt.plot(fuji[:, 0], fuji[:, 3]) 
plt.show()
```

### [Problem 2] Create a function to calculate the slope of a point

```python
def slope(array, pos):    
    bt_diff = fuji[1:,3] - fuji[:-1,3]
    at_diff = fuji[1:,0] - fuji[:-1,0]
    gradient = bt_diff/at_diff
    return gradient[pos-1]
slope_fuji_at136 = slope(fuji, 136)
print("Gradient at point 136 is {}".format(slope_fuji_at136))
```

### [Problem 3] Create a function to calculate the destination

```python
def compute_next_postion(pos, elevations, alpha = 0.2):
    destination_point = pos-alpha*slope(fuji, pos)
    destination_point = np.maximum(1, destination_point)
    destination_point = np.minimum(len(elevations)-1, destination_point)
    destination_point_int = np.round(destination_point).astype(np.int) 
    return destination_point_int
fuji_elevations = fuji[:,3]
print("Next postion after {} is {}".format(136, compute_next_postion(136, fuji_elevations, 0.2)))
```

### [Problem 4] Creating a function to go down the mountain

```python
def goingdown(start_pos, elevations, alpha ):
    position = [start_pos]
    while position[-1] not in position[:-1]:
        position.append(compute_next_postion(position[-1], elevations, alpha = 0.2))
    return position[:-1]
fuji_elevations = fuji[:,3]
print("Descending list from {} position: {}".format(136, goingdown(136, fuji_elevations, alpha = 0.2)))
```

### Problem 5] Visualization of the descent process

```python
fuji_elevations = fuji[:, 3]
descended_list = goingdown(136, fuji_elevations, alpha = 0.2)
descended_elevations = fuji_elevations[descended_list]

plt.title('Downhill process of Mt. Fuji')
plt.ylabel('elevation[m]')
plt.xlabel('position')
plt.plot(fuji_elevations)
plt.scatter(descended_list, descended_elevations, 20, color='red')
plt.show()
```

```python
fuji_elevations = fuji[:, 3]
descended_elevations = fuji_elevations[descended_list]

plt.title('Going down Mt. Fuji')
plt.ylabel('elevation[m]')
plt.xlabel('number of iterations')
plt.plot(descended_elevations)
plt.show()
```

### [Problem 6] Change of initial value

```python
goingdown_list = []
for i, start in enumerate (range(1,300)):
    goingdown_list.append(goingdown(start, fuji_elevations, alpha = 0.2))

plt.title('Downhill process of Mt. Fuji for initial position at 219')
plt.ylabel('elevation[m]')
plt.xlabel('position')
plt.plot(fuji_elevations)
plt.scatter(goingdown_list[219], fuji_elevations[goingdown_list[219]], 20, color='red')
plt.show()
```

### [Problem 7] (Advance problem) Hyperparameter change

```python
start = 136
goingdown_list_alpha = []
alpha_list = np.array([0.001, 0.01, 0.1, 0.5, 1])
for i, a in enumerate (alpha_list):
    goingdown_list_alpha.append(goingdown(start, fuji[:, 3], a))
```
