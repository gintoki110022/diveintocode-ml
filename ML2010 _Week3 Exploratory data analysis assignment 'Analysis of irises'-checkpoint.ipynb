### [Problem 6] Creating a diagram
Make a pie chart of the number of samples per label (while showing percentages)

```python
counts = total["Species"].value_counts()
plt.title("Analysis of irises")
plt.pie(counts, labels=iris_data.target_names, autopct='%1.1f%%', shadow=True, startangle=90)
plt.show()
```

Select features one by one and visualize the distribution of data for each label using a box plot

```python
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
for i in features:
    plt.boxplot([total[total["Species"] == 0][i], total[total["Species"] == 1][i], total[total["Species"] == 2][i]])
    plt.xticks([1, 2, 3], iris_data.target_names)
    plt.title("Analysis of irises {}".format(i))
    plt.xlabel('Species')
    plt.ylabel('cm')
    plt.show()
```

Select features one by one and visualize the distribution of data for each label using a violin plot

```python
features = ["sepal_length", "sepal_width", "petal_length", "petal_width"]
for i in features:
    plt.violinplot([total[total["Species"] == 0][i], total[total["Species"] == 1][i], total[total["Species"] == 2][i]])
    plt.xticks([1, 2, 3], iris_data.target_names)
    plt.title("Analysis of irises {}".format(i))
    plt.xlabel('Species')
    plt.ylabel('cm')
    plt.show()
```

With the added density information, violin plot nicely reveal the structure in the data, while a boxplot does not. And this is why violin plot is better than boxplot, when you have enough data to estimate the density.

### [Problem 7] Confirming the relationship between features

Make scatter plots of all combinations (six), such as choosing two from four features, color-coded by type.
The following figure is an example of one of them.

```python
import itertools
combine = list(itertools.combinations(total.columns[:4], 2))
figure, axis = plt.subplots(6,figsize=(5,30))
for i, com in enumerate(combine):
    axis[i].set_title(com[0] + ' and ' +com[1])
    axis[i].set(xlabel=com[0] + 'cm',ylabel=com[1] +'cm')
    axis[i].scatter(total[total['Species']==0][com[0]], total[total['Species']==0][com[1]], c='blue', label = load_iris().target_names[0])
    axis[i].scatter(total[total['Species']==1][com[0]], total[total['Species']==1][com[1]], c='red', label = load_iris().target_names[1])
    axis[i].scatter(total[total['Species']==2][com[0]], total[total['Species']==2][com[1]], c='yellow', label = load_iris().target_names[2])
    axis[i].legend()
plt.show()
```

Create a scatterplot matrix that displays scatterplots for all combinations at once
