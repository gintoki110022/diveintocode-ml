[Problem 1] Linear function


```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
a = np.arange(-50.0,50.1,0.1)
b = 1/2*a+1
print(a)
print(b)
```

[Problem 2] Array combination

```python
at = a[np.newaxis].T
bt = b[np.newaxis].T
c = np.concatenate((at,bt),axis=1)
print(c)
print(c.shape)
```

[Problem 3] Find the gradient

```python
bt_diff = c[1:,1] - c[:-1,1]
at_diff = c[1:,0] - c[:-1,0]
gradient = bt_diff/at_diff
print(gradient)
print(gradient.shape)
```

[Problem 4] Draw a graph

```python
plt.title("y = 1/2*x + 1") 
plt.xlabel("x") 
plt.ylabel("y") 
plt.plot(a[:], b[:], color ="red") 
plt.show()

plt.title("Gradient") 
plt.xlabel("x") 
plt.ylabel("Gradient") 
plt.plot(c[:-1,0], gradient, color ="red") 
plt.show()
```

[Problem 5] Python functionalization

```python
def function1(array_x):
    array_y = np.power(array_x, 2)
    return array_y
def function2(array_x):
    array_y = 2*np.power(array_x, 2) + np.power(2, array_x)
    return array_y
def function3(array_x):
    array_y = np.sin(np.power(array_x, 1/3))
    return array_y
def compute_gradient(string, function, x_range=(-50, 50.1, 0.1)):
    """
    Calculate the gradient using the amount of change.
    Parameters
    ----------------
    function : function
      The function you want to find, the one that returns the ndarray of y when you put the ndarray of x.
    x_range : tuple
      Specify the range in the same way as the argument of np.arange ().
    Returns
    ----------------
    array_xy : ndarray, shape(n, 2)
      A combination of x and y. n depends on x_range.
    gradient : ndarray, shape(n-1,)
      Function gradient. One factor is reduced to make a difference
    """
    array_x = np.arange(*x_range)
    array_y = function(array_x)
    xt = array_x[np.newaxis].T
    yt = array_y[np.newaxis].T
    c = np.concatenate((xt,yt),axis=1)
    yt_diff = c[1:,1] - c[:-1,1]
    xt_diff = c[1:,0] - c[:-1,0]
    gradient = yt_diff/xt_diff
    
    plt.title("{}".format(string)) 
    plt.xlabel("x") 
    plt.ylabel("y") 
    plt.plot(array_x[:], array_y[:], color ="red") 
    plt.show()

    plt.title("Gradient {}".format(string)) 
    plt.xlabel("x") 
    plt.ylabel("Gradient") 
    plt.plot(c[:-1,0], gradient, color ="red") 
    plt.show()
    return c, gradient
array_xy1, gradient1 = compute_gradient("y = 2*x",function1)
array_xy2, gradient2 = compute_gradient("y = 2*(x^2) +2^x ",function2)
array_xy3, gradient3 = compute_gradient("y = sin(x^(1/3))",function3,(0, 50.1, 0.1) )
```

[Problem 6] Find the minimum value

```python
print(array_xy1.min(axis=0)[1], array_xy1.argmin(axis=0)[1])
print(array_xy2.min(axis=0)[1], array_xy2.argmin(axis=0)[1])
print(array_xy3.min(axis=0)[1], array_xy3.argmin(axis=0)[1])
```
