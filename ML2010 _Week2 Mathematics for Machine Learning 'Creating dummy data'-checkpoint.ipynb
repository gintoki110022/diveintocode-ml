[Problem 1] Creating random numbers


```python
from matplotlib import pyplot as plt
%matplotlib inline
import numpy as np
```

```python
mean1 = np.array([-3, 0])
cov1 = np.array([[1.0, 0.8],[0.8, 1.0]])
np.random.seed(0)
data = np.random.multivariate_normal(mean1, cov1, 500)
print(data)
```

[Problem 2] Visualization by scatter plot

```python
plt.scatter(data[:,0], data[:,1] ,label="Scatter")
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend()
plt.show()
```

[Problem 3] Visualization by histogram

```python
plt.hist(data[:,0],60 )
plt.xlabel('x1')
plt.ylabel('frequency')
plt.title('Histogram of x1')
plt.xlim(-6, 3)
plt.ylim(0, 30)
plt.show()
```

```python
plt.hist(data[:,1],60 )
plt.xlabel('x2')
plt.ylabel('frequency')
plt.title('Histogram of x2')
plt.xlim(-6, 3)
plt.ylim(0, 30)
plt.show()
```

[Problem 4] Addition of data

```python
mean2 = np.array([0, -3])
np.random.seed(0)
data1 = np.random.multivariate_normal(mean2, cov1, 500)
plt.scatter(data[:,0], data[:,1],c="b",label="Scatter" )
plt.scatter(data1[:,0], data1[:,1],c="g",label="Scatter" )
plt.xlabel("x1")
plt.ylabel("x2")
plt.legend(('1','4'))
plt.show()
```

[Problem 5] Data combination

```python
data_total = np.concatenate([data, data1], axis=0)
print(data_total.shape)
```

[Problem 6] Labeling

```python
label_data = np.zeros((500,1))
label_data1 = np.ones((500,1))
data_after_label = np.concatenate([data, label_data], axis=1)
data1_after_label = np.concatenate([data1, label_data1], axis=1)
datatotal_after_label = np.concatenate([data_after_label, data1_after_label], axis=0)
print(datatotal_after_label)
print(datatotal_after_label.shape)
```
